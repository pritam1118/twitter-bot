```python
#Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tr=pd.read_csv("training_data.csv")
```

```python
tr.head()
```

```python
def get_heatmap(df):
    #This function gives heatmap of all blank or NaN values
    plt.figure(figsize=(10,6))
    sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
    plt.tight_layout()
    return plt.show()
```

```python
get_heatmap(tr)
```

```python
tr.shape
```

```python
tr.dropna(
    axis=0,
    how='all',
    thresh=None,
    subset=None,
    inplace=True
)
```

```python
tr.shape
```

```python
get_heatmap(tr)
```

```python
tr.isnull().sum()
```

```python
tr["location"].fillna("None", inplace = True)

tr["description"].fillna("None", inplace = True)

tr["url"].fillna("None", inplace = True)

tr["status"].fillna("None", inplace = True)

tr["has_extended_profile"].fillna("None", inplace = True)
```

```python
tr.isnull().sum()
```

```python
get_heatmap(tr)
```

```python
tr.head()
```

```python
tr.replace(to_replace="NaN",value = "None",inplace = True)
tr.replace(to_replace="null",value = "None",inplace = True)
```
